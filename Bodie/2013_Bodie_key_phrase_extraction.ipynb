```python
from io import StringIO
from bs4 import BeautifulSoup
from tika import parser
import re

file_path = "Books Index for Phrases/2013_Bodie.Kane.Marcus_Investments (McGraw-Hill, 10th Ed).pdf"

file_data = []
_buffer = StringIO()
data = parser.from_file(file_path, xmlContent=True)
xhtml_data = BeautifulSoup(data['content'])
for page, content in enumerate(xhtml_data.find_all('div', attrs={'class': 'page'})):
    print('Parsing page {} of pdf file...'.format(page+1))
    _buffer = StringIO()
    _buffer.write(str(content))
    parsed_content = parser.from_buffer(_buffer.getvalue())
    _buffer.truncate()
    file_data.append({'id': 'page_'+str(page+1), 'content': parsed_content['content']})
```

## Glossary

```python
## Glossary ## 
## Page 2 (G-1) to Page 15 (G-14)
glossary_pages = xhtml_data.find_all('div', attrs={'class': 'page'})[1: 15]
```

```python
glossary_page_1 = glossary_pages[0]
```

```python
for p in glossary_page_1.findAll(name='p'):
    # print(p.text)
    text = p.text
    ## remove page number
    if re.search(r'G-\d{1,2}', text): 
        continue

    ## remove Glossary 
    if re.search(r'Glossary', text): 
        continue

    ## remove Final PDF to printer
    if re.search(r'Final PDF to printer', text): 
        continue
    
    text = text.replace('\n', '').strip()
    text = text.split('    ')[-2:]
    
    if text != ['']:
        print(text)
```

```python
PHRASES = []
EXPLANATION = []

for page in glossary_pages:

    for p in page.findAll(name='p'):
        # print(p.text)
        text = p.text
        ## remove page number
        if re.search(r'G-\d{1,2}', text): 
            continue

        ## remove Glossary 
        if re.search(r'Glossary', text): 
            continue

        ## remove Final PDF to printer
        if re.search(r'Final PDF to printer', text): 
            continue
        
        text = text.replace('\n', '').strip()
        text = text.split('    ')[-2:]
        
        if text != ['']:
            # print(text)
            phrase, explanation = text
            phrase = phrase.replace('"', '').replace('  ', ' ')
            explanation = explanation.replace('"', '').replace('  ', ' ')
            PHRASES.append(phrase)
            EXPLANATION.append(explanation)
```

```python
import pandas as pd 
df = pd.DataFrame({'ID': [i for i in range(len(PHRASES))],
            'phrase': PHRASES,
            'textbook': ['2013_Bodie.Kane.Marcus_Investments' for _ in range(len(PHRASES))],
            'definition_or_explanations': EXPLANATION})
```

```python
df.to_csv('glossary_extraction_2013_Bodie.Kane.Marcus_Investments.csv', index=False)
```

## Subject Index

```python
## subject index ## 
## Page 19 (I-4) to Page 36 (I-21)
subject_index_pages = xhtml_data.find_all('div', attrs={'class': 'page'})[18: ]
```

```python
subject_index_pages[0]
```

```python
subject_index_pages_1 = subject_index_pages[0]
```

```python
res = []
for p in subject_index_pages_1.findAll(name='p'):
    # print(p.text)
    text = p.text
    ## remove page number
    if re.search(r'I-\d{1,2}', text): 
        continue

    # remove bod61671_sidx_I4-I22.indd   4bod61671_sidx_I4-I22.indd   4 8/2/13   7:44 AM8/2/13   7:44 AM
    if re.search(r'bod61671_sidx_I4-I22.indd', text): 
        continue

    ## remove Final PDF to printer
    if re.search(r'Final PDF to printer', text): 
        continue

    ## first word of each letter section

    if re.search('^[A-Z]', text):
        text = text.split('\n')[1].split(',')[0]

    ## regular case
    if re.search('^   [A-Z]', text):
        text = text.split(',')[0]
        try:
            text = text.split(';')[0]
        except:
            pass
        if text:
            # print(text)
            res.append(text.strip().replace('\n', ''))
```

```python
## 

res = []

for pages in subject_index_pages:
    for p in pages.findAll(name='p'):
        # print(p.text)
        text = p.text

        text = text.replace('’', "'")
        ## remove page number
        if re.search(r'I-\d{1,2}', text): 
            continue

        # remove bod61671_sidx_I4-I22.indd   4bod61671_sidx_I4-I22.indd   4 8/2/13   7:44 AM8/2/13   7:44 AM
        if re.search(r'bod61671_sidx_I4-I22.indd', text): 
            continue

        ## remove Final PDF to printer
        if re.search(r'Final PDF to printer', text): 
            continue

        ## first word of each letter section

        if re.search('^[A-Z]', text):
            text = text.split('\n')[1].split(',')[0]

        ## regular case
        if re.search('^   [A-Z]', text):
            text = text.split(',')[0]
            try:
                text = text.split(';')[0]
            except:
                pass
            if text:
                # print(text)
                res.append(text.strip().replace('\n', ''))
```
