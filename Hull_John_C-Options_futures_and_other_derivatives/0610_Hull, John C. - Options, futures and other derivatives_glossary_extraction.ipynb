```python
from io import StringIO
from bs4 import BeautifulSoup
from tika import parser
import re

file_path = "../Books Index for Phrases/(Important) Hull, John C. - Options, futures and other derivatives, 9th ed. (2013).pdf"

file_data = []
_buffer = StringIO()
data = parser.from_file(file_path, xmlContent=True)
xhtml_data = BeautifulSoup(data['content'])
for page, content in enumerate(xhtml_data.find_all('div', attrs={'class': 'page'})):
    print('Parsing page {} of pdf file...'.format(page+1))
    _buffer = StringIO()
    _buffer.write(str(content))
    parsed_content = parser.from_buffer(_buffer.getvalue())
    _buffer.truncate()
    file_data.append({'id': 'page_'+str(page+1), 'content': parsed_content['content']})
```

```python
def find_index_of_first_word_begin_with_lower_case(tokenized_text):
    token_id = -1
    flag = 0
    # print(tokenized_text)
    for id, token in enumerate(tokenized_text):
        if re.search('^[a-z]', token):
            token_id = id
            flag = 1
            break

    if flag == 0:
        return 

    if flag == 1:
        return token_id
```

```python
import nltk
PHRASES = []
EXPLANATION = []

for p in glossary_page_5.findAll(name='p'):
    # print(p.text)
    ## for each p tag, find the first word begin with lower case letter, it's token id is i
    ## key phrase token id [0: I-1]
    ## explanation id [I-1: ]
    text = p.text.replace('\n', ' ')
    text = text.replace('–', '-')
    text = text.replace('ô', 'o')
    text = text.replace('�', '')
    text = text.replace('’', "'")
    text = text.replace(" '", "'")
    text = text.replace('ð100XÞ', '(100-X)')
        

    if not text:
        continue

    if text == 'Glossary of Terms':
        continue

    if re.search('^\d{3}', text):
        continue

    tokenized_text = nltk.word_tokenize(text)
    tokenized_text_id = find_index_of_first_word_begin_with_lower_case(tokenized_text)
    if not tokenized_text_id:
        # no lower case letter word found
        match = re.search('See', text)
        if match:
            start, _ = match.span()
            phrase = text[:start-1]
            explanation = text[start:]
            PHRASES.append(phrase)
            EXPLANATION.append(explanation)
        else:
            continue

    else:
        # found lower case letter word
        # token = text[tokenized_text_id]
        # match = re.search(token, text)
        # if match:
        #     start, _ = match.span()
        #     phrase = text[:start-3]
        #     explanation = text[start-1:]
        #     PHRASES.append(phrase)
        #     EXPLANATION.append(explanation)

        phrase = ' '.join(tokenized_text[:tokenized_text_id-1])
        phrase = phrase.replace('( ', '(').replace(' )', ')').replace(' .', '.')
        explanation = ' '.join(tokenized_text[tokenized_text_id-1:])
        explanation = explanation.replace('( ', '(').replace(' )', ')').replace(' .', '.')
        PHRASES.append(phrase)
        EXPLANATION.append(explanation)
            

```

```python
TEXT = []
PHRASES = []
EXPLANATION = []

for page in glossary_pages:

    for p in page.findAll(name='p'):
        text = p.text.replace('\n', ' ')
        text = text.replace('–', '-')
        text = text.replace('ô', 'o')
        text = text.replace('�', '')
        text = text.replace('’', "'")
        text = text.replace(" '", "'")
        

        if not text:
            continue

        if 'Glossary of Terms' in text:
            continue

        if re.search('^\d{3}', text):
            continue
        
        
        tokenized_text = nltk.word_tokenize(text)
        tokenized_text_id = find_index_of_first_word_begin_with_lower_case(tokenized_text)
        if not tokenized_text_id:
            # no lower case letter word found
            match = re.search('See', text)
            if match:
                start, _ = match.span()
                phrase = text[:start-1]
                explanation = text[start:]
                PHRASES.append(phrase)
                EXPLANATION.append(explanation)
                TEXT.append(text)
            else:
                continue

        else:
            phrase = ' '.join(tokenized_text[:tokenized_text_id-1])
            phrase = phrase.replace('( ', '(').replace(' )', ')').replace(' .', '.').strip()
            explanation = ' '.join(tokenized_text[tokenized_text_id-1:])
            explanation = explanation.replace('( ', '(').replace(' )', ')').replace(' .', '.').strip()
            PHRASES.append(phrase)
            EXPLANATION.append(explanation)
            TEXT.append(text)
            
```

```python
import pandas as pd 
df = pd.DataFrame({'ID': [i for i in range(len(PHRASES))],
            'original_glossary_extracted': TEXT,
            'key_phrase': PHRASES,
            'source_book': ['Hull, John C. - Options, futures and other derivatives, 9th ed. (2013)' for _ in range(len(PHRASES))],
            'definition_or_explanation': EXPLANATION})
```

## Reformat extraction

```python
import re
def find_full_expression_and_abbreviation(text):
    text_length = len(text)
    match = re.search('\(.+\)', text) # search of bracket
    match_1 = re.search('[A-Z]{2,8}', text) # search for Capitalization
    match_2 = re.search('[A-Z]/[A-Z]', text) # search for Capitalization
    if match and (match_1 or match_2):
        content_in_bracket_length = len(match.group(0)) - 2
    else:
        if re.search('\(.+', text):
            pos = re.search('\(.+', text).span()[0]
            return [text[:pos-1]]

        else:
            return [text]

    start, end = re.search('\(.+\)', text).span()
    if content_in_bracket_length > text_length / 2:
        
        full_expression = text[start+1: end-1]
        abbreviated_expression = text[: start-1]
    else:
        full_expression = text[: start-1]
        abbreviated_expression = text[start+1: end-1]

    return [full_expression, abbreviated_expression]
```

```python
full = []
extracted_acronym_1 = []
extracted_acronym_2 = []

for phrase, definition  in zip(df['key_phrase'].values, df['definition_or_explanation'].values):
     res = find_full_expression_and_abbreviation(phrase)
     if len(res) == 1:
         full.append(res[0])
         extracted_acronym_1.append('')
         extracted_acronym_2.append('')

     if len(res) == 2:
         full.append(res[0])
         extracted_acronym_1.append(res[1])
         extracted_acronym_2.append('')
```

```python
df['key_phrase'] = full
df['extracted_acronym_1'] = extracted_acronym_1
df['extracted_acronym_2'] = extracted_acronym_2
```
